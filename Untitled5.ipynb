Hangman — Game Logic
1. Intro / Name Entry

Display a welcome message introducing the game
Call askname() — prompts the user for their name

Only accepts alphabetical characters (no spaces or special characters)
On invalid input: print "Please try again" and re-prompt
After a set number of failed attempts, exit or use a default name


Store the result as player_name


2. Game Mode Selection

Print "Hello [player_name], are you ready to play Hangman?"
Prompt: "Please select a difficulty: easy or hard"

Validate input — keep asking until they enter "easy" or "hard"
Easy mode: lives = 10
Hard mode: lives = 5




3. The Game
Setup:

Randomly select word from a predefined word list
Create hidden_word — a list of "_" for each letter (e.g. _ _ _ _ _)
Create guessed_letters = [] to track previous guesses
Print "The word has [len(word)] letters"
Print the current hangman graphic (based on lives remaining)

Each Turn:

Display current hidden_word and lives remaining
Ask: "Guess a letter, or if you're feeling confident, guess the whole word"
Player chooses one input type per turn:

Letter guess — disables word guess for that turn
Word guess — disables letter guess for that turn



If guessing a letter:

Check if letter was already in guessed_letters

If yes: "You already guessed that — no life deducted"
If no: add to guessed_letters

If letter is in word: reveal all its positions in hidden_word, print "Good guess!"
If letter is not in word: lives -= 1, print "Wrong! [lives] lives remaining", update hangman graphic




Check if hidden_word is fully revealed → player wins

If guessing the whole word:

If guess == word:

Print "You guessed it! The condemned man did not hang! You win! 🎉"
If you enjoyed the game, please play again :)"`


If guess != word:

lives -= 1, print "Wrong word! [lives] lives remaining"



When lives == 0:

Break out of the game loop
Print the completed hangman graphic
Print "You lose — unlucky. Try harder next time!"
Reveal the correct word: "The word was: [word]"


4. Play Again

After a win or loss, return to game mode selection (game_mode)
Ask "Would you like to play again? (yes/no)"

If yes → restart from step 2 (keep the same player name)
If no → print a goodbye message and exit
